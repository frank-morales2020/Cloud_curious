```python
import os
from huggingface_hub import HfApi, login
from google.colab import userdata

# 1. AUTHENTICATION
# Ensure 'HF_TOKEN' is in your Colab Secrets
try:
    HF_TOKEN = userdata.get('HF_TOKEN')
    login(token=HF_TOKEN)
    api = HfApi()
    print("✅ Authenticated successfully.")
except Exception as e:
    print(f"❌ Auth Error: {e}. Check your Colab Secrets for HF_TOKEN.")

# 2. REPOSITORY SETUP
repo_id = "frankmorales2020/gemma-4-e4b-resilient-vision"
api.create_repo(repo_id=repo_id, exist_ok=True)

# 3. DEFINE FILES TO DEPLOY
files_to_deploy = {
    "bench_vision.py": "/content/bench_vision.py",
    "requirements.txt": "/content/requirements.txt",
    "vllm_config.yaml": "/content/vllm_config.yaml"
}

# 4. UPLOAD TECHNICAL FILES
print("🚀 Uploading technical files...")
for path_in_repo, local_path in files_to_deploy.items():
    if os.path.exists(local_path):
        api.upload_file(
            path_or_fileobj=local_path,
            path_in_repo=path_in_repo,
            repo_id=repo_id,
            commit_message=f"Add {path_in_repo} for Resilient AI Challenge"
        )
        print(f"  ✓ {path_in_repo} uploaded.")
    else:
        print(f"  ⚠️ Skipping {path_in_repo}: File not found.")

# 5. GENERATE README WITH YAML METADATA
# This header fixes the "empty or missing yaml metadata" warning
readme_text = """---
license: gemma
library_name: vllm
base_model: google/gemma-4-E4B-it
tags:
- vision
- image-to-text
- resilient-ai-challenge
---

# Gemma 4 E4B Resilient Vision

This model is a compressed version of Gemma 4 optimized for the Resilient AI Challenge 2026 Image-to-Text category.

## Optimization & Inference
- **Inference Engine**: vLLM (v0.20.2)
- **Compression**: Native vLLM default optimization and memory management.
- **Precision**: 4-bit (via vLLM handshake).

## Evaluation Setup
The model is served via vLLM with challenge-mandated parameters:
- **Temperature**: 1.0
- **Top-P**: 0.95
- **Top-K**: 64

## License
Original Google Gemma License.
"""

# Upload the finalized README
api.upload_file(
    path_or_fileobj=readme_text.encode(),
    path_in_repo="README.md",
    repo_id=repo_id,
    commit_message="Fix YAML metadata and finalize README"
)

print(f"\n✅ Deployment complete! Repo: https://huggingface.co/{repo_id}")
print("\nNEXT STEP: Manually paste your benchmark logs into the README via the HF website.")
```
